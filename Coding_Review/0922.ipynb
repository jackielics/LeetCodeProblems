给定整数n和k，返回[1, n]中**字典序**第k小的数字

```python
def solution(n:int, k: int):
	'''
	Solution1: use built-in .sort()
	Time: O(nlogn) -> O(logn)
	Space: O(n)
	Trie
	'''
	# dict order: sort by str(), about the quantity??? 
	# Assume small and No Bug
	# [1, n].str
	return int(sorted([str(x) for x in range(1, n + 1)])[k - 1])

# solution(13, 2) # output: 10
# solution(1, 1) # output: 1

```
